```python
df_new = df[(df.price.quantile(0.001)<df.price)&(df.price.quantile(0.999)>df.price)]
```

```python
df_new['zscore'] = (df_new.price - df.price.mean())/df.price.std()
```

```python
df_new.zscore.min()
```

```python
df_new = df_new[(df_new.zscore>-0.65) & (df_new.zscore>.65)]
```
